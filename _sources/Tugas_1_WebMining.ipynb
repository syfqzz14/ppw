```python
import requests
import pandas as pd
from google.colab import files  # untuk download CSV

# API Springer
API_KEY = "99d0b72ecaa3adcb617eb084ed046cf2"
URL = "http://api.springernature.com/meta/v2/json"

# Kata kunci pencarian
keywords = ["web mining", "web usage mining", "data mining"]

# List untuk menampung semua hasil
all_records = []

for keyword in keywords:
    params = {
        "q": keyword,
        "api_key": API_KEY,
        "p": 20   # jumlah record per query
    }

    response = requests.get(URL, params=params)

    if response.status_code == 200:
        data = response.json()
        print(f"Kata kunci '{keyword}' - total hasil: {data['result'][0]['total']}")

        for record in data["records"]:
            all_records.append({
                "keyword": keyword,
                "doi": record.get("doi", "N/A"),
                "title": record.get("title", "No title"),
                "publicationName": record.get("publicationName", "N/A"),
                "abstract": record.get("abstract", "No abstract"),
                "url": record.get("url", [{}])[0].get("value", "N/A")
            })
    else:
        print("Error:", response.status_code, response.text)

# Simpan ke CSV
df = pd.DataFrame(all_records)
csv_filename = "springer_crawling.csv"
df.to_csv(csv_filename, index=False, encoding="utf-8-sig")

print("\nData berhasil disimpan ke springer_crawling.csv")

# 🔽 Download langsung di Colab
files.download(csv_filename)
```
